```python
from gtts import gTTS
from IPython.display import Audio

text = 'Esta es una prueba de audio. Constante R es 0.082 atmosferas litro / mol Kelvin'
tts = gTTS(text=text, lang='es')
filename = "/content/voice.mp3"
tts.save(filename)

Audio("/content/voice.mp3")
```
