En la figura anterior se muestran cinco reactores interconectados, se necesitan cinco ecuaciones
de balance de masa para representar el sistema. En el reactor 1, velocidad de la
masa que entra es

$
5(10) + Q_{31}c_3
$

y la velocidad de la masa que sale es

$Q_{12}c_1 + Q_{15}c_1$

Como el sistema se encuentra en estado estacionario, los flujos de entrada y de salida
deben ser iguales:

$
5(10) + Q_{31}c_3 = Q_{12}c_1 + Q_{15}c_1
$



o, sustituyendo los valores de la figura,

$
6c_1 – c_3 = 50
$

Realice los balances de masa para los otros reactores y obtenga el sistema de ecuaciones lineales. Utilize un método numérico para resolver estas cinco ecuaciones con las
cinco incógnitas que son las concentraciones.

$$
6c_1-c_3=50\\
-3c_1+3c_2=0\\
-c_2+9c_3=160\\
-c_2-8c_3+11c_4-2c_5=0\\
-3c_1-c_2+4c_5=0
$$

```python
import numpy as np
```

```python
A=np.array([[6,0,-1,0,0],
            [-3,3,0,0,0],
            [0,-1,9,0,0],
            [0,-1,-8,11,-2],
            [-3,-1,0,0,4]])
b=np.array([50,0,160,0,0])
```

```python
x=np.linalg.solve(A,b)
print(x)
```
