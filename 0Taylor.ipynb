```python
from sympy import Symbol,cos
from sympy.plotting import plot
```

```python
x = Symbol('x')
```

```python
fx=cos(x)
print(fx)
```

```python
t = fx.series(x,0,7).removeO()
print(t)
```

```python
plot(fx,t,(x,-4,4));
```
