```python
# prompt: crear una grafica en 3d de la funcion del sombrero y de x^2+y^2 usando plotly

import plotly.graph_objects as go
import numpy as np

# Create the x and y coordinates
x = np.arange(-2, 2, 0.1)
y = np.arange(-2, 2, 0.1)
X, Y = np.meshgrid(x, y)

# Calculate the sombrero function
Z1 = np.sin(np.sqrt(X**2 + Y**2)) / np.sqrt(X**2 + Y**2)

# Calculate the x^2 + y^2 function
Z2 = X**2 + Y**2 - 1

# Create the 3D surfaces
fig = go.Figure()
fig.add_trace(go.Surface(z=Z1, x=X, y=Y, name='Sombrero Function', colorscale='Viridis'))
fig.add_trace(go.Surface(z=Z2, x=X, y=Y, name='x^2 + y^2 - 1', opacity=0.7, colorscale='Hot'))

# Update layout
fig.update_layout(title='Sombrero Function and x^2 + y^2', autosize=False,
                  width=1000, height=800,
                  margin=dict(l=65, r=50, b=65, t=90))

# Show the plot
fig.show()
```
