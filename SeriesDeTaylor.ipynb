#Series de Taylor

Se construye a partir de la serie infinita
$$
f(x)=f(x_0)+f'(x_0)(x-x_0)+\frac{f''(x_0)}{2!}(x-x_0)^2+\frac{f'''(x_0)}{3!}(x-x_0)^3+\dots
$$

Obtener la serie truncada de ***Maclaurin*** de $f(x)=cos(x)$ con $x_0=0$ de $8^{vo}$ orden

```
import sympy as sp #importa el paquete de matematica simbolica
import numpy as np #importa el paquete de arreglos
import matplotlib.pyplot as plt #importa el paquete de graficos
```

```
x0=1 #punto origen
orden=9 #orden de la serie
x=sp.Symbol('x') #se crea la variable simbolica
serie=sp.ln(x).series(x,x0,orden).removeO() #obtiene la serie de Taylor
print(serie)
```

```
s=sp.lambdify(x,serie,'numpy') #conviente la serie a matematica numerica
xs=np.linspace(x0-0.5,x0+1.5) #se crea el vector de las x's
ys=np.log(xs) #se evalua la funcion con las x's
yst=s(xs) #se evalua la serie con las x's
```

```
plt.figure(figsize=(15,8)) #tamano de la grafica
plt.plot(xs,ys,xs,yst) #crea la grafica
plt.grid() #dibuja la rejilla
plt.title('Serie de Taylor de $f(x)=cos(x)$') #titulo de grafica
plt.legend(['$f(x)=Ln(x)$','Serie de Taylor']) #leyendas de las grafcias
plt.xlabel('x') #titulo de eje x
plt.ylabel('y') #titulo de eje y
plt.show() #muestra grafica
```
